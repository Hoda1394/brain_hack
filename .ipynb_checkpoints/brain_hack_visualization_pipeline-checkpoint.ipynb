```python
import nobrainer
import nibabel as nib
import nilearn
import matplotlib.pyplot as plt
from nilearn import plotting
import matplotlib.cm as cm
```

```python
#!pip install --no-cache-dir nobrainer
```

```python
# download a sample dataset from nobrainer
csv_path = nobrainer.utils.get_data()
filepaths = nobrainer.io.read_csv(csv_path)
```

```python
filepaths
```

```python
# take first image and label as sample
image_path = filepaths[0][0]
label_path = filepaths[0][1]
```

```python
image_path
```

```python
label_path
```

```python
# load image and segmentation with nibabel
image = nib.load(image_path)
seg = nib.load(label_path)
```

```python
# plot the image
# Original anatomy
fig1 = plt.figure(figsize=(15,6))
plotting.plot_anat(image,
                       cmap=cm.gist_gray,
                       display_mode='mosaic',
                       cut_coords= 7,
                       black_bg =True,
                       dim= -0.5,
                       figure=fig1
                       )
```

```python
# plot the segmentation
```

```python
# plot the image overlayed with segmentation
# freesurfer label
fig2 = plt.figure(figsize=(15,6))
plotting.plot_roi(seg,
                      bg_img=image,
                      figure=fig2,
                      display_mode='mosaic',
                      cut_coords=7 ,
                      black_bg =True,
                      dim= -0.5,
                      threshold=0.001,
                      #vmax=2,
                      )
```
